```python
import pandas as pd

df = pd.read_csv("ES (1).csv")
```

```python
#add moving averages to the data frame using a moving average cross over

hours_per_day = 7

df['9-day'] = df['close'].rolling(9*hours_per_day).mean() #add moving averages to the dataframe
df["21-day"] = df['close'].rolling(21*hours_per_day).mean() #one trading month

df[0:10]
```

```python
#add a signal column, tells you when you should be long or short
#long -> predict market trends up; 1 means go long next bar
    # you buy because you expect the price to rise
#short -> predict market trends down; -1 means go short next bar
    # you sell first because you expect the price to fall then buy back later at a lower price to pocket the difference
# 0 means stay flat and take no position
import numpy as np 

df["signal"] = np.where(df["9-day"] > df["21-day"], 1, 0)
df["signal"] = np.where(df["9-day"] < df["21-day"], -1, df["signal"])
df.dropna(inplace=True)
df.head()
```

```python
#calculate instantaneous returns
df["return"] = np.log(df["close"]).diff()
df["system_return"] = df["signal"] * df["return"]
df["entry"] = df.signal.diff()
```
